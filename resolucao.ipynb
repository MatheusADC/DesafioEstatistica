```python
# Bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Dicionário de faturamento
dict_faturamento = {
    'data_ref': [
        '2023-01-01', 
        '2020-02-01', 
        '2021-03-01', 
        '2022-04-01', 
        '2023-05-01',
        '2023-06-01', 
        '2020-07-01', 
        '2021-08-01', 
        '2022-09-01', 
        '2023-10-01',
        '2022-11-01', 
        '2023-12-01',
        ],
    'valor': [
        400000, 
        890000, 
        760000, 
        430000, 
        920000,
        340000, 
        800000, 
        500000, 
        200000, 
        900000,
        570000, 
        995000,
        ]
}
```

```python
# Tratamento
df_faturamento = pd.DataFrame.from_dict(dict_faturamento)
df_faturamento
```

```python
# Média de faturamento
media_faturamento = round(df_faturamento['valor'].mean(), 2)
media_faturamento
```

```python
# Gráfico de Barras Vertical
plt.bar(dict_faturamento['data_ref'], dict_faturamento['valor'], color='blue', alpha=0.7)
plt.xticks(rotation=90) 
plt.show()
```

```python
# Gráfico de Linhas
plt.plot(dict_faturamento['data_ref'], dict_faturamento['valor'], marker='o', color='green')
plt.xticks(rotation=65) 
plt.show()
```
